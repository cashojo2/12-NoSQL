## Part 3: Exploratory Analysis

### 1. Establishments that have a hygiene score equal to 20

```python
# Query to find the establishments with a hygiene score of 20:
query = {'scores.Hygiene': 20}

# Cast the results as a list and save to a variable
results = list(establishments.find(query))

# #of documents in the result:
count = establishments.count_documents(query)
print(f'Count: {count}')

# display first document:
print('First Result:')
print('----------------------')
pprint(results[0])
```

```python
import pandas as pd

# convert to df
df = pd.DataFrame(results)

# display results
row_count = df.shape[0]
print(f'Number of rows: {row_count}')
df.head(10)
```

### 2. Establishments in London that have a `RatingValue` greater than or equal to 4

```python
# Query the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4:

# query = {'LocalAuthorityName': 'London','RatingValue': {'$gte': 4}}

query = {
    '$or': [
        {'AddressLine3': 'London'},
        {'AddressLine4': 'London'}
    ],
    'RatingValue': {'$gte': 4}
}


# cast results as list and save to variable
results = list(establishments.find(query))

# display results
count = establishments.count_documents(query)
print(f'Count: {count}')

print('First Result:')
print('----------------------')
pprint(results)


# group_query = {'$group': {'_id': '$LocalAuthorityName'}
               
```

```python
# convert results to Pandas df
df = pd.DataFrame(results)

row_count = df.shape[0]
print(f'Number of rows: {row_count}')
df.head(10)
```

### 3. Top 5 establishments with a Rating of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"

```python
query = {'BusinessName': 'Penang Flavours'}
results = list(establishments.find(query))
pprint(results)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.485459
longitude = 0.008496

query = {
    '$and': [
        {'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search}},
        {'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search}},
        {'RatingValue': 5}
    ]
}

sort = [('scores.Hygiene', 1)]

# Cast the results as a list and save the results to a variable
results = list(establishments.find(query).sort(sort))

# Display the number of documents in the result
result_count = len(results)
print(f'Number of documents within the specified range: {result_count}')

# Display the first few documents in the results
pprint(results[:5])
```

```python
# convert results to Pandas df
df = pd.DataFrame(results)

print(f'Number of rows: {row_count}')
row_count = df.shape[0]
df.head(10)
```

### 4. Number of establishments in each Local Authority area that have a hygiene score of 0 using pipeline

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest


# Build the aggregation pipeline
# 1. Match query to find only establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 20}}


# 2. Aggregate by Local Authority
group_query = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}

# 3. Sort highest to lowest
# Create a dictionary that will allow the pipeline to sort by count in descending order
sort_values = {'$sort': { 'count': -1 }}

# Assemble pipeline
pipeline = [match_query, group_query, sort_values]

# Execute
results = list(establishments.aggregate(pipeline))

# Print the first 10 results
pprint(results)
```

```python
# Convert to a df
df = pd.DataFrame(results)

# Display results
row_count = df.shape[0]
print(row_count)
df.head(10)
```
